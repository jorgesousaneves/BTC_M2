```python
%sql
CREATE OR REPLACE TABLE btc_m2.gold.analise_correlacao AS
SELECT
  data,
  preco_btc,
  liquidez_global,
  (preco_btc / first(preco_btc) OVER (ORDER BY data)) * 100 as indice_btc,
  (liquidez_global / first(liquidez_global) OVER (ORDER BY data)) * 100 as indice_liquidez
FROM btc_m2.silver.consolidado
ORDER BY data
```

```python
%sql
OPTIMIZE btc_m2.gold.analise_correlacao
ZORDER BY (data)
```
